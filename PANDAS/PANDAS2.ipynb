# RESAMPLING PANDAS TIME SERIES

Resampling: statistical method over different time intervals
    
    Need string to specify frequency like "M"= month or "A"= year
    
Downsampling: reduce date time  rows to slower frequency like from daily to weekly
    
Upsampling: increase data time rows to faster frequency like from daily to hourly

Interpolate: Interpolate values according to different methods like "linear", "time" or "index"

```python
data2.resample("A").mean() #Tüm yıllara göre değerlerin ortalması alınır iki 92 yılı iki 93 yılı var
```

```python
data2.resample("M").mean() #Var olan aylara göre en erken tarih ve en geç tarih arasında tüm aylara göre bir dizlim
```

Numeric olan NaN değerleri interpolate etmek istersem; yani lineer bir şekilde arasını doldurmak.( Tarisel bir şekilde ilerlerken. )

```python
data2.resample("M").first().interpolate("linear")
```

```python
#Mean ile interpolate yapabilirizz.
data2.resample("M").mean().interpolate("linear")
```

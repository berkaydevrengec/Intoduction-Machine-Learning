# Building Data Frames From Scratch

We can build data frames csv and we did earlier.

Also we can build dataframe from dictionaries

zip() method : This function returns a list of tuples. Where the i-th tuple contains the i-th element from each of the argument sequence or iterables.

Adding new columns

Broadcasting : Create a new column and assing a value to entire column

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
country= ["Spain", "France"]
population= ["11","22"]
list_label= ["country", "population"]
list_col= [country, population]

zipped_list= list(zip(list_label, list_col))

data_dict= dict(zipped_list)
df=pd.DataFrame(data_dict)
df
```

```python
df["capital"]=["madrid","paris"]  #yeni sütun yaratma
df
```

```python
df["income"]=0 #yeni sütun yaratma ve 0 atama 
df
```

# VISUAL EXPLORATORY DATA ANALYSIS

Plot

Subplot

Histogram:

    bins: number of bins
    
    range(tuble): min and max values of bins
    
    normed(boolean): normalize or not
    
    cumulative(boolean): compute cumulative distribution
        


```python
data= pd.read_csv("pokemon.csv")
data.head()
```

```python
#Plotting all data

data1= data.loc[:,["Attack","Defense","Speed"]]
data1.plot()
```

Karmaşıklıktan kurtulabilmek için subplot kullanalım

```python
data1.plot(subplots=True)
plt.show()
```

```python
data1.plot(kind= "scatter", x="Attack", y="Defense") #Scatter nokta nokta
```

```python
data1.plot(kind= "hist", y="Defense")
```

```python
data1.plot(kind="hist",y= "Defense", bins=50, range= (0,250))
```

```python
fig, axes= plt.subplots(nrows=2, ncols=1)
data1.plot(kind="hist", y= "Defense", bins=50, range= (0,250),  ax= axes[0])
data1.plot(kind="hist", y= "Defense", bins=50, range= (0,250),ax= axes[1],cumulative= plt.savefig("graph.png"))
plt
```

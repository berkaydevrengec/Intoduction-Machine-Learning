# Cost Function

Her bir image için forward propagation adımını tamamlanmasının ardından bunun bize olan maaliyet hesabı. Her bir image için z değerleri yani weight ardından bias değerleri ile toplanıp daha sonra sigmoid fonksiyonun içine soktuktan sonra elde edilen y_head değerleri. Bilindiği üzere olduşma fazla işlem yükü olacağından (her image için hesap yapmak), tüm girdilerim olan imageleri bir array içine koymak da maliyetimizi düşürecektir. Tüm bir matrisi bir kerede forward propagatiion adımına sokup 348 adet loss değeri elde edip toplamdaki cost değerimi (maliyet, tutar) elde ediyorum. Cost değerimin yüksek çıkması modelimin başarısızlığını ortaya koyar.

# Backward Propagation


Başlangıçta weight ve bias değerlerine yanlış değerler atarsak, modelimiz yüksek bir cost değeri elde eder. Bu nedenle, weight ve bias değerlerini, cost fonksiyonunu azaltacak şekilde güncellememiz gerekiyor. Bu güncelleme işlemi Backward Propagation adı verilen süreçle gerçekleştirilir. Backward Propagation, cost fonksiyonundan başlayarak geriye doğru gidilerek weight ve bias değerlerinin güncellenmesini sağlayan bir yöntemdir.

Bu güncelleme işlemini gerçekleştirirken Gradient Descent adlı bir yöntem kullanılır. Gradient Descent, cost fonksiyonunun minimum değerine ulaşmak için iteratif olarak weight ve bias değerlerini günceller. Bu işlemde, cost fonksiyonunun gradienti kullanılarak, weight ve bias değerleri gradientin zıt yönde olduğu yöne doğru güncellenir. Böylece, cost fonksiyonu en aza indirgenerek modelimiz daha iyi bir performans sergiler.


Weight ve Bias Güncellenmesi:

Weight= Weight- Learning Rate*Cost'un Weight'e Göre Türevi

Learning Rate el ile ayarlanan bir değer. Her seferinde yazıp denenmesi gerekir.

# Optimization Algorithm with Gradient Descent

En uygun değerleri bulmak için, dereceli optimizasyon adımları.Optimize etmek istediğimiz şey, cost fonksiyonunun değerinin minimum olacağı weight ve bias değerlerini bulmaktır. Bu sayede, modelimiz en iyi performansı sergileyebilir.

Hataların toplamı, yaptığımız tahminlerin doğruluğunu yansıtır. Eğer tahminlerimiz yanlış ise, hata fonksiyonunun değeri yüksek olur. Tam tersi durumda, doğru tahminler yaparsak hata fonksiyonunun değeri düşer.

Cost fonksiyonunun değeri, weight ve bias parametrelerine bağlıdır. Bu nedenle, cost fonksiyonunu azaltmak için weight ve bias değerlerini güncellemek gerekir.

Modelimiz, weight ve bias parametrelerini öğrenerek, cost fonksiyonunu minimize eden en iyi değerleri bulmalıdır. Bu sayede, modelimiz daha iyi tahminler yapabilir ve daha iyi performans sergileyebilir.

Eğer son güncellemede weight değeri değişmemişse, cost fonksiyonunun minimum noktasına ulaşmışız demektir.

Türev, bir fonksiyonun o noktaya göre eğimini gösterir. Dolayısıyla, minimum cost fonksiyonu değerini bulmak için o fonksiyonun weight'e göre türevini alıp, sıfıra eşitlemeliyiz. Eğer cost fonksiyonunun minimum değerindeysek, türevin değeri sıfırdır.

Forward propagation metodunun içerisinde z değerini bulacağız. Ayrıca birde sigmoid function'u bulmalıyız. Sigmoid function metodunu yazalım.



```python
# calculation of z
#z = np.dot(w.T,x_train)+b
def sigmoid(z):
    y_head = 1/(1+np.exp(-z))
    return y_head
```

# Implementing Forward Propagation


```python
# Forward propagation steps:
# find z = w.T*x+b
# y_head = sigmoid(z)
# loss(error) = loss(y,y_head)
# cost = sum(loss)
def forward_propagation(w,b,x_train,y_train):
    z = np.dot(w.T,x_train) + b
    y_head = sigmoid(z) # probabilistic 0-1
    loss = -y_train*np.log(y_head)-(1-y_train)*np.log(1-y_head)
    cost = (np.sum(loss))/x_train.shape[1]      # x_train.shape[1]  is for scaling
    return cost 
```

Artık bir cost değeri elde ettik. Şimdi sırada backward propagation var. Yani weight ve bias değerlerini güncellemek.

# Implementing Backward Propagation


```python
def forward_backward_propagation(w,b,x_train,y_train):
    # forward propagation
    z = np.dot(w.T,x_train) + b
    y_head = sigmoid(z)
    loss = -y_train*np.log(y_head)-(1-y_train)*np.log(1-y_head)
    cost = (np.sum(loss))/x_train.shape[1]      # x_train.shape[1]  is for scaling
    # backward propagation
    derivative_weight = (np.dot(x_train,((y_head-y_train).T)))/x_train.shape[1] # x_train.shape[1]  is for scaling
    derivative_bias = np.sum(y_head-y_train)/x_train.shape[1]                 # x_train.shape[1]  is for scaling
    gradients = {"derivative_weight": derivative_weight,"derivative_bias": derivative_bias}
    return cost,gradients

```

# Implementing Update Parameters


Bu fonksiyon, parametreleri (w ve b) güncellemek için geriye doğru yayılımı ve maliyet hesaplama işlemlerini kullanır. Bu işlemler için forward_backward_propagation fonksiyonunu kullanır. Öncelikle forward_backward_propagation fonksiyonunu kullanarak maliyeti ve gradyanları hesaplar. Sonra, belirtilen sayıda iterasyon boyunca güncelleme işlemini gerçekleştirir. Güncelleme işlemi, w ve b'yi, öğrenme oranı ve gradyanlar kullanarak bir önceki aşamaya göre günceller. Her 10 iterasyonda bir, maliyeti ve iterasyon sayısını ekrana yazdırır. Son olarak, güncellenmiş parametreleri (w ve b), gradyanları ve maliyet listesini döndürür. Ayrıca, iterasyon sayısına karşı maliyeti gösteren bir grafik oluşturur.





```python
# Updating(learning) parameters
def update(w, b, x_train, y_train, learning_rate,number_of_iterarion):
    cost_list = []
    cost_list2 = []
    index = []
    # updating(learning) parameters is number_of_iterarion times
    for i in range(number_of_iterarion):
        # make forward and backward propagation and find cost and gradients
        cost,gradients = forward_backward_propagation(w,b,x_train,y_train)
        cost_list.append(cost)
        # lets update
        w = w - learning_rate * gradients["derivative_weight"]
        b = b - learning_rate * gradients["derivative_bias"]
        if i % 10 == 0:
            cost_list2.append(cost)
            index.append(i)
            print ("Cost after iteration %i: %f" %(i, cost))
    # we update(learn) parameters weights and bias
    parameters = {"weight": w,"bias": b}
    plt.plot(index,cost_list2)
    plt.xticks(index,rotation='vertical')
    plt.xlabel("Number of Iterarion")
    plt.ylabel("Cost")
    plt.show()
    return parameters, gradients, cost_list
#parameters, gradients, cost_list = update(w, b, x_train, y_train, learning_rate = 0.009,number_of_iterarion = 200)
```

# Implementing Prediction


Predict yapabilmek için benim moodele ihtiyacım var modelim nelerden oluşuyor. w,b, x_test değerlerimizi verdik ve x_test yani test etmek için ayırdığımız %15 lik kısımı sınava sokma zamanı. Önce z yi bulalım ardından sigmoid fonksiyonuna sokalım. Bunlar artık bu projede gündelik işlere dönüştü. artık predicte geldik. Önce y_prediction için 0 lardan oluşan bir array oluşturduk. (Bunu bir rezervasyon gibi düşünebiliriz. Eğer bunu yapmadan devam etsek kodumuz müthiş yavaş çalışacaktır. Boşa başımıza bela almayalım.) Tüm z yi dolşaıp threshold değerimize uygun olan etiketleri yapıştırmasını rica ediyoruz.

```python
# prediction
def predict(w,b,x_test):
    # x_test is a input for forward propagation
    z = sigmoid(np.dot(w.T,x_test)+b)
    Y_prediction = np.zeros((1,x_test.shape[1]))
    # if z is bigger than 0.5, our prediction is sign one (y_head=1),
    # if z is smaller than 0.5, our prediction is sign zero (y_head=0),
    for i in range(z.shape[1]):
        if z[0,i]<= 0.5:
            Y_prediction[0,i] = 0
        else:
            Y_prediction[0,i] = 1

    return Y_prediction
# predict(parameters["weight"],parameters["bias"],x_test)
```

learning_rate : öğrenme hızım, num_iteration: ne kadar update yani kaç kere türev alacağımızdır.

```python
def logistic_regression(x_train, y_train, x_test, y_test, learning_rate ,  num_iterations):
    # initialize
    dimension =  x_train.shape[0]  # that is 4096
    w,b = initialize_weights_and_bias(dimension)
    # do not change learning rate
    parameters, gradients, cost_list = update(w, b, x_train, y_train, learning_rate,num_iterations)
    
    y_prediction_test = predict(parameters["weight"],parameters["bias"],x_test)
    y_prediction_train = predict(parameters["weight"],parameters["bias"],x_train)

    # Print train/test Errors
    print("train accuracy: {} %".format(100 - np.mean(np.abs(y_prediction_train - y_train)) * 100))
    print("test accuracy: {} %".format(100 - np.mean(np.abs(y_prediction_test - y_test)) * 100))
    
logistic_regression(x_train, y_train, x_test, y_test,learning_rate = 0.01, num_iterations = 150)
```

Correlation Map: It is the relationship that allows us to understand the relationship between the features. Correlation is a coefficient that takes a value between 1 and -1 and indicates the relationship between two variables.


(Correlation Map: Featurelar arsındaki ilişkiyi anlamamızı sağlayan ilişkidir. Korelasyon bilindiği üzere 1 ve -1 arasında değer alan ve iki değişken arasındaki ilişkiyi belirten bir katsayıdır.)

Seaborn: It is a visualization library and we draw with the heatmap() function of this library. We insert the correlation results of our data (data.corr()) into the HeatMap function.

(Seaborn: Görselleştirme kütüphanesidir ve bu kütüphanenin heatmap() fonksiyonu ile çizdirme yapıyoruz. HeatMap fonksiyonu içine verimizin (data.corr()) korelasyon sonuçlarını veririz.)

sns.heatmap(data.corr(),annot=True, linewidth=.5,fmt=" .1f",ax=ax)

The heatmap takes the correlation table, annot=True to make the correlation values appear in the map, linewidth=.5 to specify the plots of the table, fmt=.1f to write only one digit after zero.

(Heatmap korelasyon tablosunu alır, annot=True korelasyon değerlerinin map içinde görünmesini sağlamak için, linewidth=.5 tablonun çizlerinin belirlenmesini, fmt=.1f sıfırdan sonra yalnızca bir basamak yazmasını sağlar.)

```python
f,ax= plt.subplots(figsize=(18,18))
sns.heatmap(data.corr(),annot=True, linewidth=.5,fmt=" .1f",ax=ax)
```

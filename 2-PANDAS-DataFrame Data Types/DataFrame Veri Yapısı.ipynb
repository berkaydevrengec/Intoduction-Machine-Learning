# DATAFRAME VERİ YAPISI

```python
import pandas as pd
```

Series veri yapısının yalnızca bir sütundan oluştuğunu öğrendik, şimdi ise DataFrame yapısını ele alacağız. DataFrame satır ve sütunlardan meydana gelir.

```python
data= {"Name":["John","Hamlet","Pearl","Love","Nick","Micheal","Robert","William"],
      "Score":[90,25,45,86,79,13,49,89],
      "Sports":["Swim","Football","Basketball","Handball","Taekwondo","Ski","Tennis","karate"],
      "Genre":["M","M","F","F","M","M","M","M"]}
df=pd.DataFrame(data)
```

```python
df
```

ilk beş satırı görmek için head methodunu kullanabiliriz.

```python
df.head()
```

Son beş satırı görmek için tail() methodunu kullanabiliriz. Aynı zamanda parantezler içine yazdığımız sayı kadar satır görmek de mümkündür.

```python
df.tail()
```

```python
df.tail(2)
```

Sütun sıralarını değiştirmek de mümkündür. Örneğin;

```python
df=pd.DataFrame(data,columns=["Name","Sports","Genre","Score"])
df
```

Eğer sütunda olmayan bir sütun adı girmeye çalışılırsa eksik veri olarak görülür. Aşağıda görüldüğü üzere Age sütunun tüm değerleri boştur.

```python
df=pd.DataFrame(data,columns=["Name","Sports","Genre","Score","Age"])
df
```

İndexlere etiket atamak da mümkündür.

```python
df=pd.DataFrame(data,columns=["Name","Sports","Genre","Score","Age"],
                index=["one","two","three","four","five","six","seven","eight"])
df
```

İstenilen sütunu ekrana yazdırmak istersek;

```python
df["Sports"]
```

Birden fazla sütun istersek önce istenilen sütunları bir değişkene atıyoruz.

```python
columns=["Name","Sports"]
df[columns]
```

İstenilen sütuna . aracılığıyla ulaşmak da mümkündür.

```python
df.Sports
```

Belirli bir satırı seçmek için loc niteliğinden faydalanabiliriz.

```python
df.loc[["one"]]
```

```python
df.loc[["one","eight"]]
```

Aynı zamanda değerleri olmayan yaş sütununa değer atamasını gerçekleştirelim.

```python
df["Age"]=18
```

```python
df
```

Aynı uzunlukta bir listeyi yaş sütununa atamayı da gerçekleştirebiliriz.

```python
Ages=[17,18,19,45,24,16,30,27]
df["Age"]=Ages
df
```

İstenilen bir sütunu silmek istersek del anahtar kelimesini kullanabiliriz.

```python
df["Pass"]=df.Score>50
df
```

```python
del df["Pass"]
df
```

SÖZLÜK YAPISI İLE DATAFRAME OLUŞTURMA                                                                                                
Bilindiği üzere sözlük yapısı key:value şeklindedir. Önce bir sözlük oluşturalım ardından dataframe yapısına çevirelim.

```python
notes= {"Math":{"John":60,"Pearl":50,"Nick":85},
        "Physics":{"John":90,"Pearl":30,"Nick":0}}
df2=pd.DataFrame(notes)
df2

```

DateFrame in satır ve sütunları yer değiştirebilir. Transpoz işlemi.

```python
df2.T
```

Satır ve sütunlara isim verilebilir.

```python
df2.index.name="Name"
df2.columns.name="Course"
df2
```

Values niteliği DataFrame veri yapısını iki boyutlu bir diziye dönüştürür.

```python
df2.values
```

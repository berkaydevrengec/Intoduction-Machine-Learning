# TIDY AND PIVOTING DATA

We tidy data with melt(). We look melt() func:

```python
import pandas as pd
```

```python
data= pd.read_csv("pokemon.csv")
```

```python
data.head()
```

```python
#lets melt
#id_vars = what we do not wish to melt
#value_vars= what we want to melt

melted= pd.melt(frame=data, id_vars= "Name", value_vars=["Attack", "Defense"])
melted
```

Bu kod, "pandas" kütüphanesindeki "melt()" fonksiyonunu kullanarak "data" adlı veri çerçevesinin ("dataframe") "Name" sütununu sabit tutarak "Attack" ve "Defense" sütunlarını tek bir sütuna ("variable") dönüştürür ve bu sütunun değerlerini ("value") de bir başka sütuna ekler. Böylece "melted" adında yeni bir veri çerçevesi oluşturulur.

Örneğin, "data" veri çerçevesinde her satır bir Pokemon'a ait bilgileri içeriyorsa, "melted" veri çerçevesinde her satır bir Pokemon'un adını, bir statünün ("Attack" veya "Defense") adını ve o statünün değerini içerecektir. Bu, veri çerçevesinin "uzun format" haline getirilmesine olanak tanır ve bazı analiz veya görselleştirme işlemleri için daha uygun bir yapı sunabilir. (CHATGPT)

# PIVOTING DATA

Reverse melting

```python
#Index is name
#I want to make that columns are variable
#Finally values in columns are value

melted.pivot(index="Name", columns="variable", values="value")
```
